#### Question (1)
- Find the number of Gold, Silver, and Bronze won by the Women in Gymnastics sport.

```python
#Import Pandas
import pandas as pd
```

```python
#Loading the csv.file
df = pd.read_csv('/Users\FELIX\DaSciPySQL\Data\Olympics_data\olympics.csv')
df
```

```python
#Renaming column to make it easier to read the data
df.columns = ['City','Edition','Sport','Discipline','Athlete','NOC','Gender','Events','Event_gender','Medal']
df
```

```python
#Create a list in other to filter the required categories
gen = ['Women','Gymnastics']

#Checking the required Columns in the created list
filt = df['Gender'].isin(gen) & df['Sport'].isin(gen)

#Printing a statement for the result
print("The table below contains the number of Gold, Silver, and Bronze won by the Women in Gymnastics sport :\n" )

#locating the Medal column using the defined filter
pd.DataFrame(df.loc[filt, 'Medal'].value_counts())
```

#### Question (2)
- Find the number of Men who lives in Los Angeles and in Football Discipline.

```python
#Create a list in other to filter the required categories (Men)
gen = ['Men']

#Checking the required Columns if they meet the requirement in  the question 
filt2 = (df['NOC'] == 'Los Angeles') & (df['Discipline'] == 'Football')

#Creating a variable to locate, check and sum the requirement in the Gender column using the defined filter
num_men = df.loc[filt2, 'Gender'].isin(gen).sum()
num_men

#Setting a conditionaL statement incase we have no result
if num_men == 0:
    print("There isn't a record of Men who live in Los Angeles and are in Football Discipline.")
else:   
    print('There are',num_men,"Men who live in Los Angeles and are in Football Discipline.")
```
